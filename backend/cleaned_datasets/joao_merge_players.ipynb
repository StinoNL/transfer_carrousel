```python
import pandas as pd
import os

# Load the dataset
df_players = pd.read_csv("clean_data/clean_players.csv")

# Show first rows
df_players.head()
```

```python
# age in 2025

df_players['age'] = df_players['date_of_birth'].apply(lambda x: 2025 - pd.to_datetime(x).year)
```

```python
# keeping only relevant columns
df_filtered_players = df_players[['player_id', 'name', 'age', 'position','market_value_in_eur']]

df_filtered_players.head()
```

```python
import os

# create the directory
os.makedirs("to_merge_data", exist_ok=True)
```

```python
df_filtered_players.to_csv("to_merge_data/filtered_players.csv", index=False)
```
