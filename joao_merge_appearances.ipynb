```python
import pandas as pd
import os

# Load the dataset
df_appearances = pd.read_csv("clean_data/j_clean_appearances.csv")

# Show first rows
df_appearances.head()
```

```python
# filter for the 2024/2025 season
df_filtered_appearances = df_appearances[df_appearances['date'] > '2024-08-01']

df_filtered_appearances.head()
```

```python
# drop unnecessary columns
df_filtered_appearances = df_filtered_appearances.drop(['appearance_id', 'game_id', 'competition_id', 'yellow_cards', 'red_cards', 'goals', 'assists'], axis=1)

df_filtered_appearances.head()
```

```python
# rename columns for merge
df_filtered_appearances = df_filtered_appearances.rename(columns={'player_club_id': 'club_id'})
```

```python
#grouping the data
df_filtered_appearances = df_filtered_appearances.groupby(['player_id', 'player_name', 'club_id'])['minutes_played'].sum().reset_index()

df_filtered_appearances.head()
```

```python
import os

# create the directory
os.makedirs("to_merge_data", exist_ok=True)
```

```python
df_filtered_appearances.to_csv("to_merge_data/filtered_appearances.csv", index=False)
```
